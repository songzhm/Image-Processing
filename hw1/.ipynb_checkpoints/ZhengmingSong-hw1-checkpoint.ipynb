```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def GammaCorrection(r, gamma):
    return 255*(r/255)**gamma
```

```python
gammas = [0.04, 0.1, 0.2, 0.4, 0.61, 1, 1.5, 2.5, 5, 10, 25]
X = np.linspace(0,250,500,endpoint=True)
data = [[GammaCorrection(r, gamma) for r in X] for gamma in gammas]
```

```python
[plt.plot(X,y, label='gamma='+ str(gamma)) for y,gamma in zip(data,gammas)]
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5),
          fancybox=True, shadow=True, ncol=1)
plt.show()
```

```python
from PIL import Image
import scipy.misc
from matplotlib import cm
```

```python
a = Image.open('Math386-HW1-Images/image-airport.tif')
```

```python
a.show()
```

```python
I = scipy.misc.fromimage(a)
(M,N) = I.shape
```

```python
plt.figure(1,figsize=(M/100*1.5,N/100*1.5))
fig1 = plt.imshow(I,cmap=cm.gray,vmin=0,vmax=255)
```

```python
def GammaCorrection(r, gamma):
    s = round(255*(r/255)**gamma,0)
    if s>255:
        s=255
    return s
J = np.array([[GammaCorrection(x,2) for x in y] for y in I])
```

```python
plt.figure(1,figsize=(M/100*1.5,N/100*1.5))
fig1 = plt.imshow(J,cmap=cm.gray,vmin=0,vmax=255)
```
